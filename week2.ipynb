```python
import pandas as pd
df1=pd.read_csv("^spx_d (2).csv",index_col="Date")
df1.index=pd.to_datetime(df1.index,errors="coerce")
df2=pd.read_csv("^ndq_d (2).csv",index_col="Date")
df2.index=pd.to_datetime(df2.index,errors="coerce")
display(df1.describe())
display(df2.describe())
```

并排展示两个discribe输出

```python
desc1=df1.describe()
desc2=df2.describe()
desc1.columns=[col+'_spx'for col in desc1.columns]
desc2.columns=[col+'_ndq'for col in desc2.columns]
combined=pd.concat([desc1,desc2],axis=1)
print(combined)
```

横向拼接

```python
desc1.index=[idx+'_spx'for idx in desc1.index]
desc2.index=[idx+'_ndq'for idx in desc2.index]
combined=pd.concat([desc1,desc2],axis=0)
print(combined)
```

```python
display(df1.info())
display(df2.info())
```

```python
common_start=max(df1.index.min(),df2.index.min())
common_end=min(df1.index.max(),df2.index.max())
print(f"共同时间区间:{common_start}至{common_end}")
```

```python
df1_common=df1[(df1.index>=common_start)&(df1.index<=common_end)]
df2_common=df2[(df2.index>=common_start)&(df2.index<=common_end)]
display(df1_common.info())
display(df2_common.info())
```
